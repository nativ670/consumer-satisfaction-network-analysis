```python
import pandas as pd
import torch
import logging

# GitHub
from src.data_preprocessing import load_and_clean_data
from src.nlp_extraction import extract_aspects_and_sentiments
from src.network_builder import build_and_analyze_network
from src.modeling import compare_models

logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

SAVE_PATH_DATA = '/content/drive/MyDrive/Seminar_Amazon_Reviews_Processed_10k.csv'

def run_large_scale_analysis():
    # 1. Check GPU Status
    if torch.cuda.is_available():
        logger.info(f"🚀 Running on GPU: {torch.cuda.get_device_name(0)}")
    else:
        logger.warning("⚠️ Running on CPU! Go to Runtime -> Change runtime type -> Hardware accelerator -> T4 GPU")
        # if no GPU
        logger.warning("Reducing sample size to 100 to prevent CPU timeout.")
        sample_size = 100
        return

    sample_size = 10000 # batch size

    try:
        # 2. Load Data
        logger.info(f"Loading {sample_size} reviews...")
        df = load_and_clean_data().head(sample_size)

        # 3. Run NLP Extraction
        logger.info("Starting heavy NLP extraction... (This may take a few minutes even on GPU)")
        processed_df = extract_aspects_and_sentiments(df)

        # 4. Save to Drive (Checkpoint)
        processed_df.to_csv(SAVE_PATH_DATA, index=False)
        logger.info(f"✅ Extracted data safely saved to Google Drive at: {SAVE_PATH_DATA}")

        # 5. Build Network
        logger.info("Constructing Partial Correlation Network...")
        network_results = build_and_analyze_network(processed_df)
        G = network_results['graph']

        # 6. Run Statistical Modeling
        logger.info("Executing Predictive Models and Cross-Validation...")
        compare_models(processed_df, G)

        logger.info("🎉 Large-scale analysis completed successfully!")

    except Exception as e:
        logger.error(f"Execution failed: {e}")
        import traceback
        traceback.print_exc()

# run function
run_large_scale_analysis()
```
